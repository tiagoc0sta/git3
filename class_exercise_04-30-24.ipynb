1- Write a Python program to find those numbers which are divisible by 7 and multiple of 5, between 1500 and 2700 (both included).

```python
for num in range(1500, 2701):
    if num % 7 == 0 and num % 5 == 0:
        print(num)
```

2 - Write a Python program to guess a number between 1 to 9.

```python
import random

# Generate a random number between 1 and 9
random_number = random.randint(1, 9)

while True:
    # Ask the user to guess the number
    guess = int(input("Guess a number between 1 and 9: "))
    
    # Check if the guess is correct
    if guess == random_number:
        print("Congratulations! You guessed the number.")
        break
    elif guess < random_number:
        print("Too low. Try again.")
    else:
        print("Too high. Try again.")
```

3- Write a Python program to construct the following pattern, using a nested for loop.

```python
for i in range(5):
    for j in range(i):
        print("*", end=" ")
    print()

for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

4- Write a Python program to check whether an alphabet is a vowel or consonant 

```python
# Function to check if a character is a vowel
def is_vowel(char):
    vowels = ['a', 'e', 'i', 'o', 'u']
    return char.lower() in vowels

# Input from user
char = input("Enter a character: ")

# Check if the input is a single alphabet
if len(char) == 1 and char.isalpha():
    if is_vowel(char):
        print(f"{char} is a vowel.")
    else:
        print(f"{char} is a consonant.")
else:
    print("Please enter a single alphabet.")
```
